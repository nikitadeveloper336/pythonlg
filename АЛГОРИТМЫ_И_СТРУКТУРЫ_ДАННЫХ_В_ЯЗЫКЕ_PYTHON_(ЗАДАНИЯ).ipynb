Для нахождения площади треугольника со сторонами  $a$, $b$, $c$ использовать формулу Герона: $S = \sqrt{p \cdot (p - a) \cdot (p - b) \cdot (p - c)}$, где $p$ — полупериметр.


```python
from math import dist
x1 = float(input())
y1 = float(input())
x2 = float(input())
y2 = float(input())
x3 = float(input())
y3 = float(input())

a = dist((x1,y1), (x2,y2))
b = dist((x2,y2),(x3,y3))
c = dist((x1,y1),(x3, y3))
p = (a+b+c)/2
S = (p*(p-a)*(p-b)*(p-c))**0.5
S
```
